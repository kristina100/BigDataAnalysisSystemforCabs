```python
import pandas as pd
data = pd.read_csv("flow_data_weekday.csv")
```

```python
labels = data['Area']
labels
```

```python
_data = data['W']
import time
import re
print(_data)
# data_1 = _data.index=["".join(i.strip().split(':')) for i in _data]
# print(data_1)
real_data = []
for real in _data:
#     real = real[10::]
    # 
    real = re.sub('\W+', '',real)
    real_data.append(real)
print(real_data)
```

## MLPRegressor

MLPRegressor




```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
```

```python
vec_data = np.array(real_data).reshape(len(real_data),1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(vec_data, labels,random_state=1)
```

```python
print(type(y_test))
```

```python
X_train, X_test, y_train, y_test
```

```python
mlp = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_test)
```

```python
y_pred = np.array(y_pred)
# 
y_pred = np.around(y_pred)
```

```python
mlp.score(X_test, y_test)
```
