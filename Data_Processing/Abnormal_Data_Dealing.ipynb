### 提取计价器K值异常的数据

```python
k_index_under = []
k_index_over = []
k_values = data['METER_K_VALUE'].values.tolist()
for i in range(len(k_values)):
    if k_values[i] > 1380:
        k_index_over.append(i)
    if k_values[i] < 520:
        k_index_under.append(i)
```

```python
k_under = []
for i in k_index_under:
    k_under.append(data[i:i+1])
k_over = []
for i in k_index_over:
    k_over.append(data[i:i+1])
Kunder = pd.concat([k for k in k_under], axis = 0, ignore_index=True)
Kover = pd.concat([k for k in k_over], axis = 0, ignore_index=True)
```

```python
use = ['PLATE_NO', 'GET_ON_LONGITUDE', 'GET_ON_LATITUDE', 'GET_OFF_LONGITUDE', 'GET_OFF_LATITUDE', 'LOG_TIME']
temp_1 = Kunder['METER_K_VALUE'].values.tolist()
temp_2 = Kover['METER_K_VALUE'].values.tolist()
Kunder = Kunder[use]
Kover = Kover[use]
```

```python
abs_conditions = []
for i in range(len(temp_1)):
    string = '计价器K值过小，K值为' + str(temp_1[i])
    abs_conditions.append(string)
Kunder['Abnormal_Condition'] = abs_conditions
abs_conditions = []
for i in range(len(temp_2)):
    string = '计价器K值过大，K值为' + str(temp_2[i])
    abs_conditions.append(string)
Kover['Abnormal_Condition'] = abs_conditions
```

```python
Kunder = Kunder.drop_duplicates(['PLATE_NO'], keep='first',ignore_index=True)
Kover = Kover.drop_duplicates(['PLATE_NO'], keep='first',ignore_index=True)
Kover = Kover.drop([4,5])
Kover = Kover.reset_index(drop=True)
```

```python
Kunder.to_csv('计价器K值过小.csv')
Kover.to_csv('计价器K值过大.csv')
```
